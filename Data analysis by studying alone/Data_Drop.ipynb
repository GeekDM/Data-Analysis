# Data Drop

```python
import pandas as pd
ns_df = pd.read_csv('/ns_202412.csv', low_memory=False)
ns_df.head()
```

```python
ns_book = ns_df.loc[:, '번호':'등록일자']
ns_book.head()
```

```python
print(ns_df.columns)
```

```python
print(ns_df.columns[0])
```

```python
ns_df.columns != 'Unnamed: 13'
```

```python
selected_columns = ns_df.columns != 'Unnamed: 13'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
selected_columns = ns_df.columns != '부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

```python
ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis=1)
ns_book.head()
```

```python
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

```python
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```

- 판다스에서 **NaN**은 누락된 값, 비어있는 값을 의미한다.
- drop() 메서드에서 axis 매개변수의 기본값인 0은 행을, 1은 열을 삭제한다.
- inplac=True일 때 ns_book 변수에 연결된 객체가 수정되는 것은 아니다.<br>
판다스는 내부적으로 수정된 새로운 객체를 만든 후 ns_book 변수에 연결한다.<br>
따라서 **inplace** 매개변수를 사용하더라도 성능상의 이득은 없다.
- 모든 값이 NaN인 열을 삭제하려면 dropna() 메서드에 how 매개변수를 'all'로 지정하면 된다.
