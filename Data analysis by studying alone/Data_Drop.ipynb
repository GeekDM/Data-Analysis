### 중복된 행 찾기

- 판다스 데이터프레임의 중복된 행은 **duplicated() 메서드**를 사용해 검사할 수 있다.<br>
\> 처음 행을 제외한 나머지 중복된 행을 True로 표시. 아니면 False로 표시.
- duplicated() 메서드는 기본적으로 모든 열을 기준으로 중복된 행을 찾는다.<br>
subset 매개변수를 사용하면 기준 열을 지정할 수 있다.


```python
sum(ns_book.duplicated())
```

```python
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN']))
```

```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

### 그룹별로 모으기

- 같은 데이터를 하나로 합칠 때 사용하는 메서드는 **groupby() 메서드**다.
- **by 매개변수**에는 행을 합칠 때 기준이 되는 열을 지정한다.
- groupby() 메서드로 데이터를 합칠 때 기본적으로 지정된 열에 NaN이 있는 행은 삭제한다.<br>
따라서 NaN을 계산에 포함할지 안 할지를 고려해야 한다.

```python
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
```

```python
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False)
loan_count = group_df.sum()
```

```python
loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()
loan_count.head()
```

### 원본 데이트 업데이트하기

원본 데이터프레임에는 중복된 데이터가 있다. 따라서 데이터를 업데이트하기 전 다음과 같은 과정을 거쳐야 한다.


1.   duplicated() 메서드로 중복된 행을 True로 표시한 불리언 배열을 만든다.
2.   1번에서 구한 불리언 배열을 반전시켜 중복되지 않은 고유한 행을 True로 표시한다.
3.   2번에서 구한 불리언 배열을 사용해 원본 배열에서 고유한 행만 선택한다.



```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])  # 1. 중복된 행을 True로 표시한다.
unique_rows = ~dup_rows # 2. 불리언 배열을 반전시켜 고유한 행을 True로 표시한다.
ns_book3 = ns_book[unique_rows].copy()  # 3. 고유한 행만 선택한다.
```

```python
sum(ns_book3.duplicated(subset=['도서명', '저자', 'ISBN', '권']))
```

- 지정한 열을 인덱스로 설정할 때는 **set_index() 메서드**를 사용한다.

```python
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_book3.head()
```

- 다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트할 때는 **update() 메서드**를 사용한다.

```python
ns_book3.update(loan_count)
ns_book3.head()
```

- **reset_index() 메서드**로 데이터프레임 인덱스를 재설정할 수 있다.

```python
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

### 일괄 처리 함수 만들기

```python
def data_cleaning(filename):
  """
  남산 도서관 장서 CSV 데이터 전처리 함수
  :param filename: CSV 파일 이름
  """
  # 파일을 데이터프레임으로 읽는다.
  ns_df = pd.read_csv(filename, low_memory=False)

  # NaN인 열을 삭제한다.
  ns_book = ns_df.dropna(axis=1, how='all')

  # 대출건수를 합치기 위해 필요한 행만 추출하여 count_df 데이터프레임을 만든다.
  count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]

  # 도서명, 저자, ISBN, 권을 기준으로 대출건수를 groupby 한다.
  loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()

  # 원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사한다.
  dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy()

  # 도서명, 저자 ISBN, 권을 인덱스로 설정한다.
  ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)

  # loan_count에 저장된 누적 대출건수를 업데이트한다.
  ns_book3.update(loan_count)

  # 인덱스를 재설정한다.
  ns_book4 = ns_book3.reset_index()

  # 원본 데이터프레임의 열 순서로 변경한다.
  ns_book4 = ns_book4[ns_book.columns]
  return ns_book4
```

```python
new_ns_book4 = data_cleaning('/ns_202412.csv')
ns_book4.equals(new_ns_book4)
```

# 마무리

- **데이터 정제**는 수집된 데이터에서 잘못된 부분을 고치거나 제거해 필요한 데이터를 준비하는 과정이다.<br>
\> 데이터가 잘못 정제되면 분석된 결과를 왜곡시킬 수 있다.
- 데이터를 정제하는 과정과 데이터 분석 및 머신러닝에 적합한 형태로 데이터를 변환하는 과정을 합쳐 <br>
  **데이터 랭글링**(or **데이터 먼징**)이라 한다.
- 판다스의 데이터프레임과 인덱스를 하나의 값과 비교하면 데이터프레임과 인덱스에 있는 모든 원소를 비교한다.<br>
이를 **원소별 비교**라고 부른다. 비교 결과는 **불리언 배열**로 반환된다.
- **넘파이**는 파이썬의 대표적인 다차원 배열이다.<br> 판다스의 데이터프레임과 달리 한 종류의 데이터만 담을 수 있지만 매우 효율적이고 성능이 높다.<br>
파이썬의 다른 과학 패키지와 호환성이 높으며 기본 데이터 구조로 널리 사용된다.

|함수/메서드|기능|
|:--|:--|
|DataFrame.drop()|데이터프레임의 행이나 열을 삭제한다.|
|DataFrame.dropna()|누락된 값이 포함된 행이나 열을 삭제한다.|
|DataFrame.duplicated()|중복된 행을 찾아 불리언 값으로 표시한 배열을 반환한다.|
|DataFrame.groupby()|데이터프레임의 행을 그룹으로 모은다.|
|DataFrame.sum()|행 또는 열을 기준으로 합계를 계산한다.|
|DataFrame.set_index()|지정한 열을 인덱스로 설정한다.|
|DataFrame.reset_index()|데이터프레임의 인덱스를 재설정한다.|
|DataFrame.update()|다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트한다.<br>다른 데이터 프레임에 있는 NaN은 업데이트에서 제외한다.|
|DataFrame.equals()|다른 데이터프레임과 동일한 원소를 가졌는지 비교한다.<br>두 데이터프레임이 동일하면 True, 그렇지 않으면 False를 반환한다.|
